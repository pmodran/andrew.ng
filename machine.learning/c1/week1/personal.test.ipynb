```python
import numpy as np
from matplotlib import pyplot
```

```python
x_train = np.array([72, 109, 43, 150])
y_train = np.array([500, 700, 350, 900])
```

```python
m = len(x_train)
m
```

```python
pyplot.scatter(x_train,y_train)
pyplot.xlabel('Metri Patrati(m2)')
pyplot.ylabel('Pret brut($)')
pyplot.title('Preturi Imobiliare')
```

```python
w = 1  # weight
b = 1   #bias

# f(x) = wx + b  //  f(1) = 1
```

```python

def compute_linear_regression(x,w,b):
    m = len(x)
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w * x[i] + b
    return f_wb

```

```python
w1 = 5

b1 = 150

f1 = compute_linear_regression(x_train, w1, b1)

pyplot.plot(x_train, f1)
pyplot.scatter(x_train, y_train)


pyplot.title('Preturi Imobiliare')
pyplot.xlabel('m2')
pyplot.ylabel('EURO')


pyplot.show()

```

```python
x_1 = 95  # new input that we want to guess how much it costs

y_prediction = x_1*w1 + b1

print(f"{y_prediction} Euros")
```

```python
def cost_function (x, y, w, b):  # aici ne trebuie si y, deoarece vom compara f(x) care ete y predictable cu Y adevaratul
    m = len(x)
    cost_sum = 0
    for i in range(m):
        print(f'x[i]{i,x[i]}')
        fwb = ( x[i] * w ) + b # y preconizat
        cost =  ( fwb - y[i] ) ** 2
        print(f'Cost{i,cost}')
        cost_sum = cost_sum + cost
    jwb = ( 1 / ( 2 * m ) ) * cost_sum
    return jwb
    


    
```

```python

w1 = 5

b1 = 150

j1 = cost_function(x_train, y_train, w1, b1)

print (x_train)

print (y_train)

print (b1)

print (w1)
print(j1)
```
